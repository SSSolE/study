즉 각 품목당 총판매액수들을 구한 후에 액수가 가장 높은 10개의 품목을 출력하기

### [Chipotle 데이터셋의 기본 정보]

```python
# -*- coding: utf-8 -*-

import pandas as pd

# read_csv 함수로 데이터를 Dataframe 형태로 불러옵니다.
file_path = '../data/chipotle.tsv'
chipo = pd.read_csv(file_path, sep = '\t')
```

```python
print(chipo.shape)
print("------------------------------------")
print(chipo.info())
```

```python
# chipo 라는 Dataframe에서 순서대로 10개의 row 데이터를 보여줍니다.
chipo.head(10)
```

```python
print(chipo.columns)
print("------------------------------------")
print(chipo.index)
```

##### describe 함수로 요약 통계량 출력하기

```python
print(chipo.describe())
```

```python
chipo['order_id'] = chipo['order_id'].astype(str) # order_id는 숫자의 의미를 가지지 않기 때문에 str으로 변환합니다.
```

```python
print(chipo.describe()) # chipo dataframe에서 수치형 피처들의 요약 통계량을 확인합니다.
```

```python
# 가장 많이 주문한 item : top 10을 출력합니다.
item_count = chipo['item_name'].value_counts()[:10]
print(item_count)
```

### [item당 주문 개수와 총량 구하기]

```python
# item당 주문 개수를 출력합니다.
order_count = chipo.groupby('item_name')['order_id'].count()
order_count[:10] # item당 주문 개수를 출력합니다.
```

```python
# item당 주문 총량을 출력합니다.
item_quantity = chipo.groupby('item_name')['quantity'].sum()
item_quantity[:10] # item당 주문 총량을 출력합니다.
```

### [apply와 lambda 함수를 이용한 데이터 전처리]

```python
item_price의 $표시를 없애고 데이터타입을 숫자로 바꾸기 위해 전처리를 함
```

```python
print(chipo.info())
print('-------------')
chipo['item_price'].head()
```

```python
# column 단위 데이터에 apply 함수로 전처리를 적용합니다.
chipo['item_price'] = chipo['item_price'].apply(lambda x: float(x[1:]))
chipo.describe()
```

### [각 아이템의 가격 구하기]

```python
# 각 아이템의 가격을 계산합니다.
chipo_one_item = chipo[chipo.quantity == 1]
price_per_item = chipo_one_item.groupby('item_name').min()
price_per_item.sort_values(by = "item_price", ascending = False)
```

각 품목당 총 판매액수 구하기

```python
price_per_item['item_price'] # 각 아이템의 가격
```

```python
#item당 주문 총량을 출력합니다.
item_quantity = chipo.groupby('item_name')['quantity'].sum()
item_quantity
```

```python
#item당 가격을 출력합니다.
item_total_price = price_per_item['item_price'] * item_quantity
item_total_price
```

```python
#내림차순 정렬
item_total_price.sort_values(ascending = False)
```

```python
# 상위 10개만 출력하기
item_total_price.sort_values(ascending = False)[:10]
```
